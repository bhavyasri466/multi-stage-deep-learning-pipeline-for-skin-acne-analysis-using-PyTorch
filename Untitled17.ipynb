```python
import torch
import torch.nn as nn
import timm
from torch.utils.data import Dataset, DataLoader
import cv2
import os
import matplotlib.pyplot as plt

# =====================================================
# 🔶 DATASET
# =====================================================
class AcneDataset(Dataset):
    def __init__(self, root_dir, size=224):
        self.image_paths = []
        self.labels = []
        self.size = size

        self.class_map = {"Acne":0, "Clear":1, "Comedo":2}

        for class_name, label in self.class_map.items():
            folder = os.path.join(root_dir, class_name)

            print("📂 Checking:", folder)

            if not os.path.exists(folder):
                print("❌ Missing:", folder)
                continue

            for file in os.listdir(folder):
                if file.endswith((".jpg",".png",".jpeg")):
                    self.image_paths.append(os.path.join(folder,file))
                    self.labels.append(label)

        print(f"✅ Loaded {len(self.image_paths)} images from {root_dir}")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.image_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (self.size, self.size))
        img = img / 255.0

        img = torch.tensor(img).permute(2,0,1).float()
        label = torch.tensor(self.labels[idx]).long()

        return img, label


# =====================================================
# 🔶 MODEL
# =====================================================
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = timm.create_model(
            "efficientnet_b0",
            pretrained=True,
            num_classes=3
        )

    def forward(self, x):
        return self.model(x)


# =====================================================
# 🔷 TRAIN
# =====================================================
def train(model, loader, optimizer, device):
    model.train()
    total_loss = 0

    for img, label in loader:
        img, label = img.to(device), label.to(device)

        output = model(img)
        loss = nn.CrossEntropyLoss()(output, label)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)


# =====================================================
# 🔷 EVALUATE
# =====================================================
def evaluate(model, loader, device):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for img, label in loader:
            img, label = img.to(device), label.to(device)

            output = model(img)
            preds = torch.argmax(output, dim=1)

            correct += (preds == label).sum().item()
            total += label.size(0)

    acc = correct / total
    print(f"✅ Validation Accuracy: {acc:.4f}")


# =====================================================
# 🔷 SHOW PREDICTION
# =====================================================
def show_prediction(model, dataset, device):
    model.eval()

    img, label = dataset[0]
    input_img = img.unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(input_img)
        pred = torch.argmax(output, dim=1).item()

    class_names = ["Acne", "Clear", "Comedo"]

    plt.imshow(img.permute(1,2,0))
    plt.title(f"Pred: {class_names[pred]}")
    plt.axis('off')
    plt.show()


# =====================================================
# 🔷 CLASS-WISE ACCURACY
# =====================================================
def classwise_accuracy(model, loader, device):
    model.eval()
    correct = [0,0,0]
    total = [0,0,0]

    with torch.no_grad():
        for img, label in loader:
            img, label = img.to(device), label.to(device)

            output = model(img)
            preds = torch.argmax(output, dim=1)

            for i in range(len(label)):
                total[label[i]] += 1
                if preds[i] == label[i]:
                    correct[label[i]] += 1

    classes = ["Acne","Clear","Comedo"]
    for i in range(3):
        if total[i] > 0:
            print(f"{classes[i]} Accuracy: {correct[i]/total[i]:.4f}")


# =====================================================
# 🔷 MAIN
# =====================================================
if __name__ == "__main__":

    device = "cuda" if torch.cuda.is_available() else "cpu"

    train_path = "/content/Acne/Train"
    val_path = "/content/Acne/Validation"

    train_dataset = AcneDataset(train_path)
    val_dataset = AcneDataset(val_path)

    print("Train size:", len(train_dataset))
    print("Val size:", len(val_dataset))

    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    model = Model().to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)

    print("✅ Training Started...")

    for epoch in range(5):
        loss = train(model, train_loader, optimizer, device)
        print(f"Epoch {epoch+1}, Loss: {loss:.4f}")

        evaluate(model, val_loader, device)

    print("\n🔹 Class-wise Accuracy:")
    classwise_accuracy(model, val_loader, device)

    print("\n🔹 Sample Prediction:")
    show_prediction(model, val_dataset, device)
```

```python
import torch
import torch.nn as nn
import timm
from torch.utils.data import Dataset, DataLoader
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# =====================================================
# 🔶 IMPROVED MASK FUNCTION (FIXED)
# =====================================================
def generate_mask(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

    # Dual red ranges
    lower1 = np.array([0, 70, 50])
    upper1 = np.array([10, 255, 255])

    lower2 = np.array([170, 70, 50])
    upper2 = np.array([180, 255, 255])

    mask1 = cv2.inRange(hsv, lower1, upper1)
    mask2 = cv2.inRange(hsv, lower2, upper2)

    mask = mask1 + mask2

    # Morphological cleaning
    kernel = np.ones((5,5), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    # Remove small noise
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)
    clean_mask = np.zeros_like(mask)

    for i in range(1, num_labels):
        if stats[i, cv2.CC_STAT_AREA] > 150:
            clean_mask[labels == i] = 255

    return clean_mask


# =====================================================
# 🔶 DATASET
# =====================================================
class AcneDataset(Dataset):
    def __init__(self, root_dir, size=224):
        self.image_paths = []
        self.labels = []
        self.size = size

        self.class_map = {"Acne":0, "Clear":1, "Comedo":2}

        for class_name, label in self.class_map.items():
            folder = os.path.join(root_dir, class_name)

            if not os.path.exists(folder):
                continue

            for file in os.listdir(folder):
                if file.endswith((".jpg",".png",".jpeg")):
                    self.image_paths.append(os.path.join(folder,file))
                    self.labels.append(label)

        print(f"✅ Loaded {len(self.image_paths)} images from {root_dir}")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.image_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (self.size, self.size))

        # 🔥 MASK
        mask = generate_mask(img)

        # 🔥 ROI extraction
        roi = cv2.bitwise_and(img, img, mask=mask)

        img = img / 255.0
        roi = roi / 255.0

        img = torch.tensor(img).permute(2,0,1).float()
        roi = torch.tensor(roi).permute(2,0,1).float()

        label = torch.tensor(self.labels[idx]).long()

        return img, roi, label


# =====================================================
# 🔶 MODEL (DUAL INPUT + FUSION)
# =====================================================
class Model(nn.Module):
    def __init__(self):
        super().__init__()

        self.backbone1 = timm.create_model("efficientnet_b0", pretrained=True, num_classes=0)
        self.backbone2 = timm.create_model("efficientnet_b0", pretrained=True, num_classes=0)

        self.fc = nn.Sequential(
            nn.Linear(1280*2, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 3)
        )

    def forward(self, img, roi):
        f1 = self.backbone1(img)
        f2 = self.backbone2(roi)

        fused = torch.cat([f1, f2], dim=1)
        return self.fc(fused)


# =====================================================
# 🔷 TRAIN
# =====================================================
def train(model, loader, optimizer, device):
    model.train()
    total_loss = 0

    for img, roi, label in loader:
        img, roi, label = img.to(device), roi.to(device), label.to(device)

        output = model(img, roi)
        loss = nn.CrossEntropyLoss()(output, label)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)


# =====================================================
# 🔷 EVALUATE
# =====================================================
def evaluate(model, loader, device):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for img, roi, label in loader:
            img, roi, label = img.to(device), roi.to(device), label.to(device)

            output = model(img, roi)
            preds = torch.argmax(output, dim=1)

            correct += (preds == label).sum().item()
            total += label.size(0)

    acc = correct / total
    print(f"✅ Validation Accuracy: {acc:.4f}")


# =====================================================
# 🔷 VISUALIZATION (FIXED + CLEAN)
# =====================================================
def show_steps(dataset):
    img, roi, label = dataset[0]

    img_np = (img.permute(1,2,0).numpy() * 255).astype(np.uint8)
    roi_np = (roi.permute(1,2,0).numpy() * 255).astype(np.uint8)

    mask = generate_mask(img_np)

    # Overlay
    overlay = img_np.copy()
    overlay[mask > 0] = [255, 0, 0]

    blended = cv2.addWeighted(img_np, 0.7, overlay, 0.3, 0)

    plt.figure(figsize=(12,5))

    plt.subplot(1,4,1)
    plt.imshow(img_np)
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1,4,2)
    plt.imshow(mask, cmap='gray')
    plt.title("Mask (Clean)")
    plt.axis('off')

    plt.subplot(1,4,3)
    plt.imshow(roi_np)
    plt.title("ROI")
    plt.axis('off')

    plt.subplot(1,4,4)
    plt.imshow(blended)
    plt.title("Overlay (Improved)")
    plt.axis('off')

    plt.show()


# =====================================================
# 🔷 MAIN
# =====================================================
if __name__ == "__main__":

    device = "cuda" if torch.cuda.is_available() else "cpu"

    train_path = "/content/Acne/Train"
    val_path = "/content/Acne/Validation"

    train_dataset = AcneDataset(train_path)
    val_dataset = AcneDataset(val_path)

    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    print("\n🔹 Visualization of improved pipeline:")
    show_steps(train_dataset)

    model = Model().to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)

    print("\n✅ Training Started...")

    for epoch in range(5):
        loss = train(model, train_loader, optimizer, device)
        print(f"Epoch {epoch+1}, Loss: {loss:.4f}")
        evaluate(model, val_loader, device)
```

```python
# ============================
# STEP 0: UPLOAD ZIP DATASET
# ============================
from google.colab import files
uploaded = files.upload()

import zipfile
import os

zip_path = list(uploaded.keys())[0]

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content")

print("✅ Dataset extracted!")

# ============================
# FIND CORRECT PATH
# ============================
print("📂 Checking folders:")
print(os.listdir("/content"))

# 👉 CHANGE THIS AFTER SEEING OUTPUT
train_img = "/content/Acne/Train"   # <-- adjust if needed
train_mask = "/content/Acne_Masks/Train"

# ============================
# INSTALL + IMPORTS
# ============================
!pip install timm -q

import torch
import torch.nn as nn
import torch.nn.functional as F
import timm
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("✅ Imports done")

# ============================
# MASK GENERATION
# ============================
def generate_mask(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    edges = cv2.Canny(gray, 30, 100)

    mask = cv2.bitwise_or(thresh, edges)

    kernel = np.ones((3,3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    return mask

def create_masks(input_root, output_root):
    print("🔄 Generating masks...")

    if not os.path.exists(input_root):
        raise ValueError(f"❌ Path not found: {input_root}")

    for cls in os.listdir(input_root):
        in_dir = os.path.join(input_root, cls)
        out_dir = os.path.join(output_root, cls)

        os.makedirs(out_dir, exist_ok=True)

        for file in os.listdir(in_dir):
            img_path = os.path.join(in_dir, file)
            save_path = os.path.join(out_dir, file)

            img = cv2.imread(img_path)
            if img is None:
                continue

            mask = generate_mask(img)
            cv2.imwrite(save_path, mask)

    print("✅ Masks created!")

create_masks(train_img, train_mask)

# ============================
# DATASET
# ============================
class SkinDataset(Dataset):
    def __init__(self, img_root, mask_root, size=224):
        self.img_paths = []
        self.mask_paths = []
        self.labels = []
        self.size = size

        self.class_map = {"Acne":0, "Clear":1, "Comedo":2}

        for cls in self.class_map:
            img_dir = os.path.join(img_root, cls)
            mask_dir = os.path.join(mask_root, cls)

            for file in os.listdir(img_dir):
                self.img_paths.append(os.path.join(img_dir, file))
                self.mask_paths.append(os.path.join(mask_dir, file))
                self.labels.append(self.class_map[cls])

        print(f"📊 Loaded {len(self.img_paths)} samples")

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.img_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        mask = cv2.imread(self.mask_paths[idx], 0)

        img = cv2.resize(img, (224,224))
        mask = cv2.resize(mask, (224,224))

        img = img / 255.0
        mask = mask / 255.0

        img = torch.tensor(img).permute(2,0,1).float()
        mask = torch.tensor(mask).unsqueeze(0).float()

        label = torch.tensor(self.labels[idx]).long()

        return img, mask, label

dataset = SkinDataset(train_img, train_mask)
loader = DataLoader(dataset, batch_size=16, shuffle=True)

# ============================
# MODEL
# ============================
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model("efficientnet_b0", pretrained=True, num_classes=0)
        self.seg_head = nn.Conv2d(1280, 1, 1)
        self.cls_head = nn.Linear(1280, 3)

    def forward(self, x):
        feat = self.backbone.forward_features(x)
        seg = self.seg_head(feat)
        seg = F.interpolate(seg, size=(224,224), mode='bilinear')
        cls = self.backbone(x)
        return seg, cls

# ============================
# TRAIN
# ============================
def dice_loss(pred, target):
    pred = torch.sigmoid(pred)
    smooth = 1
    inter = (pred * target).sum()
    return 1 - (2*inter + smooth) / (pred.sum() + target.sum() + smooth)

device = "cuda" if torch.cuda.is_available() else "cpu"
model = Model().to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)

print("🚀 Training started...")

for epoch in range(2):
    model.train()
    total_loss = 0

    for i, (img, mask, label) in enumerate(loader):
        img, mask, label = img.to(device), mask.to(device), label.to(device)

        seg, cls = model(img)

        loss = dice_loss(seg, mask) + F.cross_entropy(cls, label)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        if i % 10 == 0:
            print(f"🔄 Epoch {epoch+1} Batch {i} Loss {loss.item():.4f}")

    print(f"📊 Epoch {epoch+1} Loss {total_loss/len(loader):.4f}")

# ============================
# RESULT
# ============================
print("✅ Training complete!")
```

```python
model.eval()

correct = 0
total = 0

with torch.no_grad():
    for img, mask, label in loader:
        img, label = img.to(device), label.to(device)

        _, cls = model(img)
        preds = torch.argmax(cls, dim=1)

        correct += (preds == label).sum().item()
        total += label.size(0)

print(f"✅ Accuracy: {correct/total:.4f}")
```

```python
import random
import matplotlib.pyplot as plt
import torch
import numpy as np

# ✅ Correct mapping from your dataset
class_names = {
    0: "Acne",
    1: "Clear",
    2: "Comedo"
}

def visualize_from_dataset(dataset, model, device):
    idx = random.randint(0, len(dataset)-1)

    img, mask, label = dataset[idx]

    # convert label safely
    label = int(label.item())

    img_input = img.unsqueeze(0).to(device)

    model.eval()
    with torch.no_grad():
        seg, cls = model(img_input)

    seg = torch.sigmoid(seg).cpu().squeeze().numpy()
    pred_mask = seg > 0.5
    pred_class = int(torch.argmax(cls, dim=1).item())

    img_np = img.permute(1,2,0).numpy()
    mask_np = mask.squeeze().numpy()

    # overlay
    overlay = img_np.copy()
    overlay[pred_mask] = [1, 0, 0]

    # -------------------------------
    # DISPLAY
    # -------------------------------
    plt.figure(figsize=(16,5))

    plt.subplot(1,5,1)
    plt.imshow(img_np)
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1,5,2)
    plt.imshow(mask_np, cmap='gray')
    plt.title("Auto Mask")
    plt.axis('off')

    plt.subplot(1,5,3)
    plt.imshow(pred_mask, cmap='gray')
    plt.title("Predicted Mask")
    plt.axis('off')

    plt.subplot(1,5,4)
    plt.imshow(overlay)
    plt.title("Acne Detection")
    plt.axis('off')

    plt.subplot(1,5,5)
    plt.imshow(img_np)
    plt.title(f"GT: {class_names[label]}\nPred: {class_names[pred_class]}")
    plt.axis('off')

    plt.show()


# 🔥 RUN MULTIPLE TIMES
for i in range(5):
    print(f"\n🔍 Sample {i+1}")
    visualize_from_dataset(dataset, model, device)
```
